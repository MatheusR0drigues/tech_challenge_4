```python
import pandas as pd
import matplotlib.pyplot as plt

# Caminho do arquivo CSV
file_path = r"G:\Meu Drive\Windows\Cursos\Ps Tech\Machine Learning Engineering\Desafio\tech_challenge_4\data\raw\AAPL_historical.csv"

# Ajustar leitura do CSV, ignorando as duas primeiras linhas
df = pd.read_csv(
    file_path,
    skiprows=2,  # Ignora as duas primeiras linhas que no fazem parte dos dados
)

# Renomear as colunas para os nomes corretos
df.columns = ["Date", "Adj Close", "Close", "High", "Low", "Open", "Volume"]

# Transformar a coluna 'Date' em ndice e converter para datetime
df['Date'] = pd.to_datetime(df['Date'])  # Converter a coluna para datetime
df.set_index('Date', inplace=True)  # Definir a coluna Date como ndice

# Exibir as primeiras linhas para verificar o ajuste
print(df.head())

# Plotar os preos de fechamento
df['Close'].plot(title="Preo de Fechamento (AAPL)")
plt.xlabel("Data")
plt.ylabel("Preo de Fechamento")
plt.show()
```
